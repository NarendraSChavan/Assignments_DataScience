**ESTIMATION AND CONFIDENCE INTERVALS**

```python
import pandas as pd
import numpy as np
from scipy import stats

durability_data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

df = pd.DataFrame({'Durability (Millions of Characters)': durability_data})
print(df)
```

```python
mean = np.mean(durability_data)
print("Sample Mean:",mean)
standard_deviation = np.std(durability_data,ddof=1)
print("Sample_Standard_deviation:",standard_deviation)
size = len(durability_data)
print("Sample_Size",size)
confidence_level = 0.99
alpha = 1 - confidence_level

degrees_of_freedom = size -1
t_critical = stats.t.ppf(1 - alpha/2, degrees_of_freedom) #determine the t_critical value

margin_of_error = t_critical * (standard_deviation / np.sqrt(size))  #margin of Error for t_critical value and standard deviation

confidence_interval_lower = df.mean().item() - margin_of_error
confidence_interval_upper = df.mean().item() + margin_of_error

print("Degrees of Freedom:",degrees_of_freedom)
print("Critical t-value:", t_critical)
print("Margin of Error:", margin_of_error)
print(f"99% Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}) millions of characters")
```

**Explaination of steps and the rationale behind using the t-distribution for this task**

1.Calculate the Sample Statistics:Firtly, Calculate the sample mean, sample standard deviation and sample size of the given data respectively.
2.Determine the confidence level and alpha: A 99% confidence level has given to determine , meaning alpha = 1 - 0.99. we need to find t-value for alpha/2 in tail.
3.Degrees of freedom:calculate the Degrees of freedom using sample size of given data
4.finding the critical value
5.calculated Margin of Error
6.Determined the Confidence Interval using Sample mean and Margin of error

**Rationale behind using the t-distribution for this task are two reasons**
1.Small Sample size:Because the sample size is 15 and which is less than 30. therefore when sample size is small t-distribution is more appropriate.
2.Sample Standard deviation: because we dont know the true population standard deviation. instead we are using the sample standard deviation ((S)) as estimate.

```python
population_std = 0.2 #Know population Standard Deviation

#z_critical value using alpha and degrees of freedom
z_critical = stats.t.ppf(1 - alpha/2, degrees_of_freedom)

margin_of_error_z = z_critical * (population_std / np.sqrt(size))

confidence_interval_lower_z = mean - margin_of_error_z
confidence_interval_upper_z = mean + margin_of_error_z

print("Known Population Standard Deviation:", population_std)
print("Z-critical value: {z_critical:.4f}")
print("Margin of Error (with known population std):",margin_of_error_z)
print(f"99% Confidence Interval (with known population std): ({confidence_interval_lower_z:.4f}, {confidence_interval_upper_z:.4f}) millions of characters")
```
