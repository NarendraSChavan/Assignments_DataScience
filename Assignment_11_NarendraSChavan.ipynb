```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("/content/sample_data/heart_disease.xlsx")
```

```python
df = pd.DataFrame(data)
```

```python
df.fillna(df['oldpeak'].mean(), inplace=True)
```

```python
#Visualization of features using Histogram
df.hist(figsize=(15,15))
plt.title('Histogram of Features')
plt.show()
```

```python
#Visualization using Box Plot
selected_columns = ['age','chol','trestbps','fbs','thalch','oldpeak','num']
for column in df[selected_columns]:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python
#visualization using correlation matrix
selected_columns = ['age','chol','trestbps','fbs','thalch','oldpeak','num']
correlation_df = df[selected_columns]
cor_matrix = correlation_df.corr()
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Selected Features')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Label Encoding is applied for sex column to determine Male(1) and Female(0)
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
```

```python
df.head(15)
```

```python
#Encoding the categorical variables
categorical_cols = ['cp', 'restecg', 'exang', 'slope', 'thal']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
#Splitting data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(df.drop('num', axis=1), df['num'], test_size=0.2, random_state=42)
```

```python
#Applying Model
Decision_tree_classifier = DecisionTreeClassifier()
```

```python
#Fit the model
Decision_tree_classifier.fit(X_train, y_train)
```

```python
#Performance testing using Evaluation Metrices
y_pred = Decision_tree_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", precision)
```

```python
recall = recall_score(y_test, y_pred,average='weighted')
print("Recall:", recall)
```

```python
F1_score = f1_score(y_test, y_pred,average='weighted')
print("F1 Score:", F1_score)
```

```python
y_pred_proba = Decision_tree_classifier.predict_proba(X_test)
roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class = 'ovr', average = 'weighted')
print("ROC AUC Score:", roc_auc)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Parameter grid
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
#Initializing GridsearchCv
grid_search = GridSearchCV(estimator=Decision_tree_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best parameters found: ", grid_search.best_params_)
print("Best score found: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
print(f"Tuned Model Accuracy: {accuracy_tuned}")
```

```python
precision_tuned = precision_score(y_test, y_pred_tuned, average='weighted')
print(f"Tuned Model Precision: {precision_tuned}")
```

```python
recall_tuned = recall_score(y_test, y_pred_tuned, average='weighted')
print(f"Tuned Model Recall: {recall_tuned}")
```

```python
f1_tuned = f1_score(y_test, y_pred_tuned, average='weighted')
print(f"Tuned Model F1 Score: {f1_tuned}")
```

```python
y_pred_proba_tuned = best_model.predict_proba(X_test)
roc_auc_tuned = roc_auc_score(y_test, y_pred_proba_tuned, multi_class='ovr', average='weighted')
print(f"Tuned Model ROC AUC Score: {roc_auc_tuned}")
```

```python
from sklearn.tree import plot_tree
#Visualization of Decision tree =
plt.figure(figsize=(20,10))
plot_tree(best_model, feature_names=X_train.columns, class_names=[str(c) for c in best_model.classes_], filled=True, rounded=True, fontsize=8)
plt.title('Tuned Decision Tree Structure')
plt.show()
```

```python
#Visualization of Decision tree using Barplot
feature_importances = pd.Series(best_model.feature_importances_, index=X_train.columns)
plt.figure(figsize=(12,6))
sns.barplot(x=feature_importances.sort_values(ascending=False).values, y=feature_importances.sort_values(ascending=False).index, palette='viridis')
plt.title('Feature Importances from Tuned Decision Tree')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

**Interview Questions:**
1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?


Some common hyperparameters of decision tree models are:
1. Maximum depth:This controls the maximum depth of the tree.
2. mininum sample split: This is the minimum number of samples required to plit an internal node.
3. mininum sample leaf: This is the minimum number of samples required to be at a leaf node.
4. criterion:This specifies the function to measure the quality of a split.
5. maximum features:This determines the number of features to consider when looking for the best split.
6. splitter:This strategy is used to choose the split at each node.


2. What is the difference between the Label encoding and One-hot encoding?

**Label Encoding:** It is suitable for ordinal categorical variables, where there is an inherent order or hierachy between the ctegories eg. Small, medium, Large could be 0,1,2.

**One-Hot Encoding:** It is ideal for nominal categorical variables, where there is no inherent order between the categories.
