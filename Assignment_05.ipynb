**EXPLORATORY DATA ANALYSIS ON A DATASET**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/sample_data/Cardiotocographic.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(df['AC'].mean(), inplace=True)
df.fillna(df['FM'].mean(), inplace=True)
df.fillna(df['UC'].mean(), inplace=True)
df.fillna(df['DL'].mean(), inplace=True)
df.fillna(df['DS'].mean(), inplace=True)
df.fillna(df['DP'].mean(), inplace=True)
df.fillna(df['ASTV'].mean(), inplace=True)
df.fillna(df['MSTV'].mean(), inplace=True)
df.fillna(df['ALTV'].mean(), inplace=True)
df.fillna(df['MLTV'].mean(), inplace=True)
df.fillna(df['Width'].mean(), inplace=True)
df.fillna(df['NSP'].mean(), inplace=True)
```

```python
df.isnull()
```

```python
print(df.describe())
```

```python
for column in df.columns:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    sns.boxplot(y=df[column])
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)

    plt.tight_layout()
    plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Cardiotocographic Dataset')
plt.show()
```
