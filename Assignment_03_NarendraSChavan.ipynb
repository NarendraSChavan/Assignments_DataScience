```python
import pandas as pd
```

```python
data = pd.read_csv("/content/sales_data_with_discounts (1).csv")
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

print('\nMedian for numerical columns:')
print(df[numerical_cols].median())

print('\nMode for numerical columns:')
for col in numerical_cols:
    modes = df[col].mode()
    if len(modes) > 1:
        print(f'{col}: Multiple modes - {modes.tolist()}')
    elif len(modes) == 1:
        print(f'{col}: {modes[0]}')
    else:
        print(f'{col}: No mode found')
```

```python
df.describe()
```

```python

import matplotlib.pyplot as plt

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col}')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
import seaborn as sns

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Create boxplots for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col])
    plt.ylabel(col)
    plt.title(f'Box Plot of {col}')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Create bar charts for each categorical column
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, palette='viridis')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title(f'Frequency of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()
```

## Key Difference between Discriptive Analysis and Data Visualization
The primary distinction between Descriptive Analysis and Data Visualization lies in their purposes. Descriptive Analysis focuses on extracting numerical insights from datasets, such as mean, median, variance, and standard deviation, to summarize and describe the data. In contrast, Data Visualization is used to represent data graphically, employing tools like histograms, box plots, and bar charts, to uncover patterns, outliers, and business insights, facilitating a deeper understanding of the data.
