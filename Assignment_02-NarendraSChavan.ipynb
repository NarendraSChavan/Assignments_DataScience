1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse

```python
lst = [10, 20, 30, 40, 50]
print("Original list:", lst)

# append
lst.append(60)
print("After append(60):", lst)

# extend
lst2 = [70, 80]
lst.extend(lst2)
print("After extend([70, 80]):", lst)

# insert
lst.insert(2, 25)
print("After insert(2, 25):", lst)

# remove
lst.remove(30)
print("After remove(30):", lst)

# pop
popped_element = lst.pop(0)
print("Popped element from index 0:", popped_element)
print("After pop(0):", lst)

# index
index_of_40 = lst.index(40)
print("Index of 40:", index_of_40)

# count
count_of_25 = lst.count(25)
print("Count of 25:", count_of_25)

# sort
lst.sort()
print("After sort():", lst)

# reverse
lst.reverse()
print("After reverse():", lst)

# clear
lst.clear()
print("After clear():", lst)
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

Data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan'],
    'Age': [24, 27, 22, 32, 24, 29, 35, 26, 30, 23],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'New York', 'Miami', 'Dallas', 'Seattle', 'Boston', 'Austin'],
    'Salary': [50000, 60000, 45000, np.nan, 50000, 70000, 80000, 55000, np.nan, 48000]
}
df = pd.DataFrame(Data)

print("Original DataFrame:")
display(df)
```

```python
selected_colums = df[['Name','City']]
print("Selected Columns\n",selected_colums)
```

```python
df.drop(['City'],axis=1,inplace=True)
print("DataFrame After Dropping City Column\n",df)
```

```python
df.isnull()
```

```python
df.fillna(df['Salary'].mean(),inplace=True)
print("DataFrame After Filling Missing Values in Salary Column\n",df)
```

```python
remove_duplicate_rows = df.drop_duplicates()
print("DataFrame After Removing Duplicate Rows\n",remove_duplicate_rows)
```
