```python
import pandas as pd
import numpy as np
```

```python
billing_data = pd.read_csv("/content/sample_data/Billing_Data.csv") #Load of data files
Patient_data = pd.read_csv("/content/sample_data/Patient_Data.csv")
```

```python
df_billing = pd.DataFrame(billing_data)   #Datframes
df_patient = pd.DataFrame(Patient_data)
```

```python
#3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
df_patient.drop('ReceptionistID', axis=1, inplace=True)
df_patient.drop('CheckInTime', axis=1, inplace=True)
```

```python
df_patient.columns
```

```python
total_bill_per_department = df_patient.groupby('Department')['BillAmount'].sum()  #4.	Use groupby to find total bill amount per department.
display(total_bill_per_department)
```

```python
df_patient.drop_duplicates(subset=['PatientID'], inplace=True)   #5.	Remove duplicate patient records based on PatientID.
```

```python
df_patient.columns
```

```python
print(df_patient)
```

```python
df_patient['BillAmount'] = df_patient['BillAmount'].replace(0.0, np.nan)   #6.	Fill missing BillAmount values with the mean bill amount.
mean_bill_amount = df_patient['BillAmount'].mean()
df_patient['BillAmount'] = df_patient['BillAmount'].fillna(mean_bill_amount)
print(df_patient)
```

```python
df_merged = pd.merge(df_patient, df_billing, on='PatientID', how='inner')  #7.	Merge the billing dataset with patient dataset on PatientID.
df_merged = pd.merge(df_patient, df_billing, on='PatientID', how='inner')
print(df_merged.head())
```

```python
#8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
new_patient_records = pd.DataFrame({
    'PatientID': [106, 107],
    'Name': ['Frank', 'Grace'],
    'Department': ['Pediatrics', 'Cardiology'],
    'Doctor': ['Dr. White', 'Dr. Smith'],
    'BillAmount': [4500.0, 7000.0],
    'InsuranceCovered': [1500, 2500],
    'FinalAmount': [3000, 4500]
})
print("New patient records created:")
print(new_patient_records)
```

```python
df_combined = pd.concat([df_merged, new_patient_records], ignore_index=True)  #9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
print("Combined DataFrame after appending new records:")
print(df_combined)
```
